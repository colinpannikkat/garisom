```python
hourly_weather = pd.read_csv("../DBG/dataset.csv")
hourly_weather = hourly_weather.dropna()
hourly_weather['year'] = hourly_weather['Year'].astype(int).astype(str)
hourly_weather['julian-day'] = hourly_weather['Day'].astype(int).astype(str)
hourly_weather['standard-time'] = hourly_weather['Hour'].astype(int).astype(str)
hourly_weather = hourly_weather.drop(['Year', 'Day', 'Hour'], axis=1)
hourly_weather = hourly_weather.set_index(['year', 'julian-day', 'standard-time'])
hourly_weather
```

```python
# Add Year column
sum_dt["Year"] = sum_dt['date'].astype(str).apply(lambda x: datetime.strptime(x, "%m/%d/%y").strftime("%Y"))
```

```python
# Add julian-day column
sum_dt["Day"] = sum_dt['date'].astype(str).apply(lambda x: datetime.strptime(x, "%m/%d/%y").strftime("%j"))
```

```python
def rename_columns(dt):
    dt = dt.rename(columns={
        'Year' : 'year', 
        'Day' : 'julian-day', 
        'am.pm' : 'standard-time', 
        'Solar_Wm2' : 'solar', 
        'Rain_mm' : 'rain', 
        'Wind_ms.1' : 'wind', 
        'Tair_C' : 'T-air',
        'Tsoil_C' : 'T-soil',
        'Tleaf (oC)' : 'leaftemp',
        'D_kPa' : 'D-MD',
        'Ymd (MPa)' : 'P-MD',
        'Ypd (MPa)' : 'P-PD',
        'E.(mmol m-2 s-1)' : 'E-MD',
        'gsw.(mmol m-2 s-1)' : 'GW',
        'K (Kg hr-1 MPa-1 m-2) ' : 'K-plant'
    })

    return dt
```

```python
sum_dt = rename_columns(sum_dt)
hourly_weather = rename_columns(hourly_weather)
```

```python
sum_dt
```

```python
# Measurement periods in the day (rounded-down and up)
time_offsets = {
    "GW": {
        "am": (7, 9),
        "pm": (15, 17)
    },
    "E-MD": {
        "am": (7, 9),
        "pm": (15, 17)
    },
    "K-plant": {
        "am": (7, 9),
        "pm": (15, 17)
    },
    "P-PD": {
        "am": (3, 5),
        "pm": (3, 5)
    },
    "P-MD": {
        "am": (13, 15),
        "pm": (13, 15)
    }
}
```

```python
keep = ['population', 'elevation', 'genotype', 'Replicate #', 'standard-time', 'date', 'treatment', 'year', 'julian-day', 'Hour']

def expand_time_ranges(df, time_offsets):
    expanded_rows = []
    for _, row in df.iterrows():
        for col, offsets in time_offsets.items():
            if col in df.columns:
                for period, (start, end) in offsets.items():
                    if row['standard-time'] == period:
                        for time in range(int(start // 1), int(end // 1 + 1) + 1):
                            new_row = row.copy()
                            for n_col in new_row.index:
                                if n_col not in keep and n_col != col:
                                    new_row[n_col] = None
                            new_row['standard-time'] = time
                            expanded_rows.append(new_row)
    return pd.DataFrame(expanded_rows)

# Apply the function to expand the time ranges
sum_dt = expand_time_ranges(sum_dt, time_offsets)
```

```python
sum_dt
```

```python
sum_dt[['julian-day', 'standard-time', 'year']] = sum_dt[['julian-day', 'standard-time', 'year']].astype(str)
```

```python
sum_dt = sum_dt.drop(columns=['genotype', 'Replicate #', 'date', 'elevation'])
ccr_pop_dt = sum_dt[sum_dt['population'] == 'CCR-COL'].groupby(['population', 'julian-day', 'standard-time', 'year']).mean(numeric_only=True)
jla_pop_dt = sum_dt[sum_dt['population'] == 'JLA-JAK'].groupby(['population', 'julian-day', 'standard-time', 'year']).mean(numeric_only=True)
nrv_pop_dt = sum_dt[sum_dt['population'] == 'NRV-NEW'].groupby(['population', 'julian-day', 'standard-time', 'year']).mean(numeric_only=True)
tsz_pop_dt = sum_dt[sum_dt['population'] == 'TSZ-SAN'].groupby(['population', 'julian-day', 'standard-time', 'year']).mean(numeric_only=True)
```

```python
ccr_pop_dt
```

```python
ccr_pop_dt = ccr_pop_dt.reset_index().set_index(['year', 'julian-day', 'standard-time']).drop(columns=['population'])
jla_pop_dt = jla_pop_dt.reset_index().set_index(['year', 'julian-day', 'standard-time']).drop(columns=['population'])
nrv_pop_dt = nrv_pop_dt.reset_index().set_index(['year', 'julian-day', 'standard-time']).drop(columns=['population'])
tsz_pop_dt = tsz_pop_dt.reset_index().set_index(['year', 'julian-day', 'standard-time']).drop(columns=['population'])
```

```python
ccr_pop_dt = pd.concat([hourly_weather, ccr_pop_dt], axis=1).reset_index()
jla_pop_dt = pd.concat([hourly_weather, jla_pop_dt], axis=1).reset_index()
nrv_pop_dt = pd.concat([hourly_weather, nrv_pop_dt], axis=1).reset_index()
tsz_pop_dt = pd.concat([hourly_weather, tsz_pop_dt], axis=1).reset_index()
```

```python
ccr_pop_dt
```

```python
# Set K in terms of basal area from leaf area
ccr_pop_dt['K-plant'] = ccr_pop_dt['K-plant'] * ccr_laperba
jla_pop_dt['K-plant'] = jla_pop_dt['K-plant'] * jla_laperba
nrv_pop_dt['K-plant'] = nrv_pop_dt['K-plant'] * nrv_laperba
tsz_pop_dt['K-plant'] = tsz_pop_dt['K-plant'] * tsz_laperba
```

```python
# Set pressures in terms of -MPa from MPa
ccr_pop_dt[['P-MD', 'P-PD']] = ccr_pop_dt[['P-MD', 'P-PD']].apply(lambda x: -x)
jla_pop_dt[['P-MD', 'P-PD']] = jla_pop_dt[['P-MD', 'P-PD']].apply(lambda x: -x)
nrv_pop_dt[['P-MD', 'P-PD']] = nrv_pop_dt[['P-MD', 'P-PD']].apply(lambda x: -x)
tsz_pop_dt[['P-MD', 'P-PD']] = tsz_pop_dt[['P-MD', 'P-PD']].apply(lambda x: -x)
```

```python
ccr_pop_dt.to_csv("ccr_hourly_data.csv", index=False)
jla_pop_dt.to_csv("jla_hourly_data.csv", index=False)
nrv_pop_dt.to_csv("nrv_hourly_data.csv", index=False)
tsz_pop_dt.to_csv("tsz_hourly_data.csv", index=False)
```
